**Split TimeStamp Column into two seperate columns**

```python
### Makinh a copy of the original dataframe before making any changes in it
df_copy = df.copy()
```

```python
### Convert 'Timestamp' column to datetime
df_copy['Timestamp'] = pd.to_datetime(df_copy['Timestamp'])

### Extract date and time into new columns
df_copy['Date'] = df_copy['Timestamp'].dt.date
df_copy['Time'] = df_copy['Timestamp'].dt.time

### Droping the original 'Timestamp' column
df_copy = df_copy.drop(columns=['Timestamp'])

print("\nDataFrame with Date and Time columns:")
print(df_copy)
```
